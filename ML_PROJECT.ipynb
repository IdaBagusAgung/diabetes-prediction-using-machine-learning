```python
from sklearn.utils import resample

# Misalkan variabel target Anda adalah 'Diabetes_binary'
majority_class = data_indikator_diabetes[data_indikator_diabetes['Diabetes_binary'] == 0]
minority_class = data_indikator_diabetes[data_indikator_diabetes['Diabetes_binary'] == 1]

# Oversample minority class
minority_oversampled = resample(minority_class, 
                                replace=True,     # sample with replacement
                                n_samples=len(majority_class),    # to match majority class
                                random_state=42)  # reproducible results

# Combine majority class with oversampled minority class
balanced_data = pd.concat([majority_class, minority_oversampled])

# Shuffle data
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Save sample to CSV
balanced_data.to_csv('DATASET/Balance-Sample-Data-Diabetes.csv', index=False)

# Verifikasi distribusi
print("Distribusi di sampel seimbang:")
print(balanced_data['Diabetes_binary'].value_counts(normalize=True))
```

```python
import pandas as pd
from sklearn.utils import resample

# Load dataset
data = pd.read_csv("DATASET/data_indikator_diabetes.csv")# Ganti dengan path ke dataset Anda

# Misalkan variabel target Anda adalah 'Diabetes_binary'
majority_class = data[data['Diabetes_binary'] == 0]
minority_class = data[data['Diabetes_binary'] == 1]

# Tentukan jumlah sampel yang diinginkan dari setiap kelas
sample_size_per_class = 5000  # Karena total sampel yang diinginkan adalah 10.000

# Undersample majority class
majority_undersampled = resample(majority_class, 
                                 replace=False,    # sample without replacement
                                 n_samples=sample_size_per_class,  # jumlah sampel yang diinginkan per kelas
                                 random_state=42)  # reproducible results

# Oversample minority class (jika perlu, tapi kali ini kita hanya membutuhkan jumlah sampel tertentu)
minority_oversampled = resample(minority_class, 
                                replace=True,     # sample with replacement
                                n_samples=sample_size_per_class,    # jumlah sampel yang diinginkan per kelas
                                random_state=42)  # reproducible results

# Combine majority class with sampled minority class
balanced_sample = pd.concat([majority_undersampled, minority_oversampled])

# Shuffle data
balanced_sample = balanced_sample.sample(frac=1, random_state=42).reset_index(drop=True)

# Save sample to CSV
balanced_sample.to_csv('sample_data_diabetes.csv', index=False)

# Verifikasi distribusi
print("Distribusi di sampel seimbang:")
print(balanced_sample['Diabetes_binary'].value_counts(normalize=True))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
data = pd.read_csv('DATASET/data_indikator_diabetes.csv')  # Ganti dengan path ke dataset Anda

# Misalkan variabel target Anda adalah 'Diabetes_binary'
X = data.drop('Diabetes_binary', axis=1)
y = data['Diabetes_binary']

# Stratified sampling dengan proporsi asli, dengan ukuran total sampel 10.000
_, X_sample, _, y_sample = train_test_split(X, y, test_size=10000/len(data), stratify=y, random_state=42)

# Gabungkan kembali fitur dan target untuk membuat sampel data lengkap
sample_data = pd.concat([X_sample, y_sample], axis=1)

# Save sample to CSV
sample_data.to_csv('sample_data_proportional.csv', index=False)

# Verifikasi distribusi
print("Distribusi di dataset asli:")
print(data['Diabetes_binary'].value_counts(normalize=True))

print("Distribusi di sampel:")
print(sample_data['Diabetes_binary'].value_counts(normalize=True))
```
